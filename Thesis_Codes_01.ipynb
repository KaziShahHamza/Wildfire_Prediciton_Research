```python
# 1. Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# 2. Load Dataset
import pandas as pd

# ðŸ”¹ Replace with your own file path inside Drive
# Example: /content/drive/MyDrive/datasets/student_performance.csv
file_path = '/content/drive/MyDrive/datasets/CA_Weather_Fire_Dataset_1984-2025.csv'

# Read the CSV file
df = pd.read_csv(file_path)

# 3. Basic Data Overview
print("Dataset Loaded Successfully!\n")

# View top rows
print("First 5 rows:\n", df.head(), "\n")

# Shape of dataset
print("Shape:", df.shape)

# Column names and data types
print("\nColumns and Data Types:\n")
print(df.dtypes)

# Basic stats for numeric columns
print("\nStatistical Summary:\n")
print(df.describe())

# Check missing values
print("\nMissing Values per Column:\n")
print(df.isnull().sum())

# Unique values in each column (useful for categorical data)
print("\nðŸ”¹ Unique Value Counts:\n")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")

# 4. Optional: Display dataframe interactively
# (Only works in Colab â€” shows scrollable table)
from IPython.display import display
display(df.head(10))
```

```python
df['PRECIPITATION'].fillna(df['PRECIPITATION'].median(), inplace=True)
df['MAX_TEMP'].fillna(df['MAX_TEMP'].median(), inplace=True)
df['MIN_TEMP'].fillna(df['MIN_TEMP'].median(), inplace=True)
df['TEMP_RANGE'].fillna(df['TEMP_RANGE'].median(), inplace=True)
df['AVG_WIND_SPEED'].fillna(df['AVG_WIND_SPEED'].mean(), inplace=True)
df['WIND_TEMP_RATIO'].fillna(df['WIND_TEMP_RATIO'].mean(), inplace=True)
```

```python
# Check missing values
print("\nMissing Values per Column:\n")
print(df.isnull().sum())
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
```

```python
# Column names and data types
print("\nColumns and Data Types:\n")
print(df.dtypes)

df['DAY'] = df['DATE'].dt.day
df['WEEK'] = df['DATE'].dt.isocalendar().week
df['QUARTER'] = df['DATE'].dt.quarter

# Display the head of the dataframe with the new columns
print("\nFirst 5 rows with new columns:\n", df.head(), "\n")
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# One-hot encode the 'SEASON' column
onehot_encoder = OneHotEncoder(sparse_output=False)
season_encoded = onehot_encoder.fit_transform(df[['SEASON']])

# Create a DataFrame with the one-hot encoded columns
season_df = pd.DataFrame(season_encoded, columns=onehot_encoder.get_feature_names_out(['SEASON']))

# Concatenate the new dataframe with the original dataframe
df = pd.concat([df, season_df], axis=1)

# Display the head of the dataframe with the new one-hot encoded columns
print("\nðŸ“Š First 5 rows with one-hot encoded season features:\n", df.head(), "\n")

# Drop the original 'SEASON' column if you no longer need it
df = df.drop('SEASON', axis=1)

# Display the columns and data types to show the changes
print("\nðŸ§¾ Columns and Data Types after one-hot encoding:\n")
print(df.dtypes)
```

```python
df['TEMP_MEAN'] = (df['MAX_TEMP'] + df['MIN_TEMP']) / 2
df['IS_RAINY'] = (df['PRECIPITATION'] > 0).astype(int)
df['WIND_CATEGORY'] = pd.cut(df['AVG_WIND_SPEED'],
                             bins=[0, 5, 10, 20, 30],
                             labels=['Low', 'Moderate', 'High', 'Extreme'])

print("\nðŸ§¾ Columns and Data Types:\n")
print(df.dtypes)


# Display the head of the dataframe with the new columns
print("\nðŸ“Š First 5 rows with new columns:\n", df.head(), "\n")
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# One-hot encode the 'WIND_CATEGORY' column
onehot_encoder = OneHotEncoder(sparse_output=False)
wind_encoded = onehot_encoder.fit_transform(df[['WIND_CATEGORY']])

# Create a DataFrame with the one-hot encoded columns
wind_df = pd.DataFrame(wind_encoded, columns=onehot_encoder.get_feature_names_out(['WIND_CATEGORY']))

# Concatenate the new dataframe with the original dataframe
df = pd.concat([df, wind_df], axis=1)

# Display the head of the dataframe with the new one-hot encoded columns
print("\nðŸ“Š First 5 rows with one-hot encoded wind category features:\n", df.head(), "\n")

# Drop the original 'WIND_CATEGORY' column if you no longer need it
df = df.drop('WIND_CATEGORY', axis=1)

# Display the columns and data types to show the changes
print("\nðŸ§¾ Columns and Data Types after one-hot encoding:\n")
print(df.dtypes)
```

```python
# Convert the boolean target feature to numerical (0 for False, 1 for True)
df['FIRE_START_DAY'] = df['FIRE_START_DAY'].astype(int)

# Separate features (X) and target (y)
X = df.drop('FIRE_START_DAY', axis=1)
y = df['FIRE_START_DAY']

# Print the shapes of X and y
print("X shape:", X.shape)
print("y shape:", y.shape)

# Show the total number of 0s and 1s
print("\nClass distribution (0: No Fire, 1: Fire):\n", y.value_counts(), "\n")

# Check class balance using a countplot
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))

# Subplot for countplot
plt.subplot(1, 2, 1)
sns.countplot(x=y)
plt.title('Class Distribution of FIRE_START_DAY (Countplot)')
plt.xlabel('FIRE_START_DAY (0: No Fire, 1: Fire)')
plt.ylabel('Count')

# Subplot for pie chart
plt.subplot(1, 2, 2)
plt.pie(y.value_counts(), labels=y.value_counts().index, autopct='%1.1f%%', startangle=90)
plt.title('Class Distribution of FIRE_START_DAY (Pie Chart)')

plt.tight_layout()
plt.show()
```
